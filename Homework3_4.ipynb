```python
import numpy
```

```python
def index_DB(D, k):
    kmers = [D[i:i+k] for i in range(0, len(D)-k+1)]
    kmer_index = {}
    for i, kmer in enumerate(kmers):
        if kmer in kmer_index:
            kmer_index[kmer].append(i)
        else:
            kmer_index[kmer] = [i]
    return kmer_index
```

```python
def seeding(D, Q, k):
    K = [Q[i:i+k] for i in range(0, len(Q)-k+1)]
    D_kmers = index_DB(D, k)

    seeds = {}
    for i, kmer in enumerate(K):
        if kmer in D_kmers:
            for j in D_kmers[kmer]:
                if kmer in seeds:
                    seeds[kmer].append((i, j))
                else:
                    seeds[kmer] = (i, j)

    return seeds
```

```python
Q = "GGATCCATTCATTA" 
D = "CTAGGATCCAGGCATACGA"
k = 4 
X = 2
```

```python
seeding(D, Q, k)
```

```python
scores = [3, -3, 2]    # мэтм, мисмэтч, гэп

def extention(D, Q, k, X, seed, scores, step):
    match, mismatch, gap = scores
    
    S_cur, S_max = 0, 0
    if step == 1:
        q, r = seed[0]+k, seed[1]+k
    if step == -1:
        q, r = seed[0]-1, seed[1]-1

    while 0 <= q < len(Q) and 0 <= r < len(D):
        S_cur += match if Q[q]==D[r] else mismatch
        if S_cur > S_max:
            S_max = S_cur
        if S_max - S_cur >= X:
            break
        q += step
        r += step

    return S_max, q
```

```python
seeds = seeding(D, Q, k)

for key, seed in seeds.items():
    score_l, left = extention(D, Q, k, X, seed, scores, -1)
    score_r, right = extention(D, Q, k, X, seed, scores, 1)
    print(f"k-мер: {key} продолжен до позиций {left+1}, {right-1}")
    
    score = score_l + score_r + k*scores[0]
    print(f"Итоговый скор = {score} для последовательности {Q[left+1:right]}\n")
```
